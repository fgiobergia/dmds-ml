```python
import pandas as pd
```

```python
df = pd.read_csv("2d-synthetic.csv")
```

```python
X = df.values[:, :2]
y = df.values[:, 2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
import numpy as np
```

```python
theta = np.pi / 4

rot = np.array([
    [ np.cos(theta), -np.sin(theta) ],
    [ np.sin(theta), np.cos(theta) ]
])
```

```python
# np.dot(X, rot)
```

```python
X_ = X @ rot
```

```python
plt.scatter(X_[:,0], X_[:, 1], c=y)
plt.axvline(7.09, c='red')
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_, y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```
